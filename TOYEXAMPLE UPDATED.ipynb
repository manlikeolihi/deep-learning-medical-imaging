```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
from segment_anything import sam_model_registry, SamPredictor

# Load SAM model
sam_checkpoint = "/Users/olihiidikwu/Desktop/segment-anything/sam_vit_b_01ec64.pth"

model_type = "vit_b"

device = "cuda" if torch.cuda.is_available() else "cpu"
sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
sam.to(device=device)
```

```python
# Load and display the sample image
image = cv2.imread("/Users/olihiidikwu/Desktop/segment-anything/sample.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Input Image")
plt.axis('off')
plt.show()
```

```python
# Initialize the predictor with the image
predictor = SamPredictor(sam)
predictor.set_image(image)

# Use a dummy foreground point â€” try different coordinates if needed
input_point = np.array([[300, 300]])  # (x, y)
input_label = np.array([1])           # 1 = foreground

# Generate masks
masks, scores, logits = predictor.predict(
    point_coords=input_point,
    point_labels=input_label,
    multimask_output=True,
)

# Display all generated masks over the image
for i, mask in enumerate(masks):
    plt.figure()
    plt.imshow(image)
    plt.imshow(mask, alpha=0.5)  # semi-transparent overlay
    plt.title(f"Mask {i+1} â€” Score: {scores[i]:.3f}")
    plt.axis('off')
    plt.show()
```
